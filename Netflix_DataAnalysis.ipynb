Creating a new column from Types column by extracting some key words


```python
def extract_genre(types):
    if isinstance(types, str):
        if 'Horror' in types:
            return 'Horror'
        elif 'Comedy' in types:
            return 'Comedy'
        elif 'Action' in types:
            return 'Action'
        elif 'Drama' in types:
            return 'Drama'
        elif 'Thriller' in types:
            return 'Thriller'
        elif 'Documentaries' in types:
            return 'Documentaries'
        elif 'Independent Movies' in types:
            return 'Independent Movies'
        elif 'Sci-Fi' in types:
            return 'Sci-Fi'
        elif 'Romantic' in types:
            return 'Romantic'
        elif 'Music' in types:
            return 'Music'
        elif 'Reality' in types:
            return 'Reality'
        elif 'Crime' in types:
            return 'Crime'
        elif 'Kids' in types:
            return 'Kids'
        elif 'Stand-Up' in types:
            return 'Stand-Up'
        elif 'Sports' in types:
            return 'Sports'
        elif 'Cult' in types:
            return 'Cult'
        elif 'International' in types:
            return 'International'
        else:
            return 'Other'
    else:
        return 'Other'

df['Genre'] = df['Type'].apply(extract_genre)
display(df[['Type', 'Genre']].head())
```

```python
df.head()
```

```python
df.groupby('Genre')['Country'].value_counts().sort_values(ascending=False).head(10)
```

```python
# Group by Genre and Country and count the occurrences, then unstack to get a matrix format
genre_country_counts = df.groupby(['Genre', 'Country']).size().unstack(fill_value=0)
genre_country_counts
```

```python
# Group by Genre and Country and count the occurrences, then unstack to get a matrix format
genre_country_counts = df.groupby(['Genre', 'Country']).size().unstack(fill_value=0)

# Select the top N genres and top M countries for better readability in the heatmap
top_genres = genre_country_counts.sum(axis=1).sort_values(ascending=False).head(10).index
top_countries = genre_country_counts.sum(axis=0).sort_values(ascending=False).head(10).index

# Filter the dataframe to include only the top genres and countries
heatmap_data = genre_country_counts.loc[top_genres, top_countries]

display(heatmap_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.heatmap(heatmap_data, annot=True, fmt='d', cmap='YlGnBu')
plt.title('Top 10 Genres by Top 10 Countries on Netflix')
plt.xlabel('Country')
plt.ylabel('Genre')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

## From the above heatmap we can obviously see that United states is leading in all Genres except for International which is bagged by Japan. Which implies that Japan is more open for International Genre releases as compared to other countries.

## Rating Descriptions
TV-Y: All children. Content is designed to be appropriate for all children.

TV-Y7: Directed to children 7 and above. May contain fantasy violence or comedic violence.

TV-Y7-FV: Directed to children 7 and above, with fantasy violence.

TV-G: General audience. Content is suitable for all ages.

TV-PG: Parental guidance suggested. May contain some material parents might find unsuitable for younger children.

PG: Parental Guidance Suggested (Movies). Some material may not be suitable for pre-teenagers.

PG-13: Parents Strongly Cautioned (Movies). Some material may be inappropriate for children under 13.

TV-14: Parents strongly cautioned. May contain some material that parents might find unsuitable for children under 14.

R: Restricted (Movies). Under 17 requires accompanying parent or adult guardian.

NC-17: No One 17 and Under Admitted (Movies).

TV-MA: Mature audience only. Content is specifically designed to be viewed by adults and may be unsuitable for children under 17.

G: General Audiences (Movies). All ages admitted.

NR / UR: Not Rated / Unrated. Content that has not been submitted for a rating, or is a director's cut/extended version that was not rated.

Unknown: This is a placeholder I used in the code for any entries where the rating information was missing from the original dataset.
